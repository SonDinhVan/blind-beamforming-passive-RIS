```python
import matplotlib.pyplot as plt
import numpy as np

def generate_black_white_table(n=8, p_black=0.3, save_pdf=False):
    """
    Generate a black & white square table with thin gaps.
    
    Args:
        n (int): size of the table (n x n).
        p_black (float): probability of a square being black.
        save_pdf (bool): if True, save as high-quality PDF.
    """
    # Generate random black (1) and white (0) squares
    board = np.random.choice([0, 1], size=(n, n), p=[1-p_black, p_black])

    # Create figure
    fig, ax = plt.subplots(figsize=(4, 4), dpi=300)
    
    # Use pcolormesh to create thin gridlines between squares
    cmap = plt.cm.gray  # black & white colormap
    ax.pcolormesh(board, cmap=cmap, edgecolors='lightgray', linewidth=0.2)

    # Remove axes
    ax.set_aspect('equal')
    ax.set_xticks([])
    ax.set_yticks([])

    plt.tight_layout()

    if save_pdf:
        plt.savefig("black_white_table.pdf", dpi=600, bbox_inches='tight')
    else:
        plt.show()

# Example usage:
generate_black_white_table(n=4, p_black=0.5, save_pdf=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def generate_all_white_table(n=8, save_pdf=False):
    """
    Generate an all-white square table with thin gaps.
    
    Args:
        n (int): size of the table (n x n).
        save_pdf (bool): if True, save as high-quality PDF.
    """
    # All squares = white
    board = np.zeros((n, n))

    fig, ax = plt.subplots(figsize=(4, 4), dpi=300)

    # White squares with thin gray gaps
    ax.pcolormesh(board, cmap="gray_r", edgecolors='lightgray', linewidth=0.2)

    # Remove axes
    ax.set_aspect('equal')
    ax.set_xticks([])
    ax.set_yticks([])

    plt.tight_layout()

    if save_pdf:
        plt.savefig("all_white_table.pdf", dpi=600, bbox_inches='tight')
    else:
        plt.show()

# Example usage:
generate_all_white_table(n=8, save_pdf=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def draw_ris_board(n=16, save_pdf=False):
    """
    Draw a RIS-like board with gradient-colored squares and white borders.
    
    Args:
        n (int): number of RIS elements per row/column (n x n).
        save_pdf (bool): if True, save as high-quality PDF.
    """
    # Create a smooth gradient across the board
    x = np.linspace(0, 1, n)
    y = np.linspace(0, 1, n)
    xv, yv = np.meshgrid(x, y)
    gradient = (xv + yv) / 2  # diagonal gradient

    fig, ax = plt.subplots(figsize=(6, 6), dpi=300)

    # Plot the gradient with white borders
    c = ax.pcolormesh(gradient, cmap="viridis", edgecolors='white', linewidth=0.5)

    # Remove axis ticks
    ax.set_aspect('equal')
    ax.set_xticks([])
    ax.set_yticks([])

    plt.tight_layout()

    if save_pdf:
        plt.savefig("RIS_board.pdf", dpi=600, bbox_inches='tight')
    else:
        plt.show()

# Example usage
draw_ris_board(n=16, save_pdf=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def draw_binary_squares(n_rows=8, n_cols=8, color="#26A69A"):
    # Generate random 0/1 values
    data = np.random.randint(0, 2, size=(n_rows, n_cols))

    fig, ax = plt.subplots(figsize=(5, 5))
    ax.set_xlim(0, n_cols)
    ax.set_ylim(0, n_rows)
    ax.set_aspect("equal")
    ax.axis("off")  # Hide axes

    # Draw squares
    for i in range(n_rows):
        for j in range(n_cols):
            rect = plt.Rectangle((j, n_rows-1-i), 1, 1, 
                                 facecolor=color, edgecolor="black", linewidth=0.8)
            ax.add_patch(rect)
            ax.text(j+0.5, n_rows-1-i+0.5, str(data[i, j]),
                    va="center", ha="center", fontsize=8, color="black")

    plt.show()

# Example: draw a 10x10 grid
draw_binary_squares(n_rows=4, n_cols=4, color="white")
```
