```python
import numpy as np
import matplotlib.pyplot as plt

from modules import node, system, sensing
from modules import optimization as opt
from configs import config

noise_power = config.NOISE_POWER
```

```python
tx = node.TX(x=0, y=0, z=10)
rx = node.RX(x=50, y=50, z=1.5)
ris = node.RIS(x=10, y=50, z=2, N=60)
sys = system.System(tx, ris, rx)

is_los = True
N_samples = 3000
n_loop = 5000

figure_name = f'fig_cdf_N{ris.N}_LOS.pdf' if is_los else f'fig_cdf_N{ris.N}_NLOS.pdf'
```

```python
all_born = []

for i in range(n_loop):
    print("---------------------------")
    print(f"Computing LOOP {i}")
    sys.gen_channels(is_los=is_los)

    # Stage 0: sending samples and collect observation:
    X = np.zeros(shape=(sys.ris.N, N_samples))
    y = np.zeros(shape=(N_samples, 1))
    for j in range(N_samples):
        # generate a random ris setting
        sys.ris.set_random_binary_phase_shifts()
        x = sys.ris.vector_x
        X[:, j] = x.flatten()
        y[j] = sys.cal_snr() - sys.c_true

    sens = sensing.Sensing(B=600, N=sys.ris.N)
    M_est, w_est, _ = sens.solve(X=X, y=y, convergence_plot=False) # Stage 1: sensing
    M_est_proj = sens.project_to_psd_lowrank(M_est, r=2) # projection to low rank
    sys.ris.vector_x = opt.solve_our_optimization(M_est_proj, w_est) # Stage 2: optimization
    sys.ris.theta = np.arccos(sys.ris.vector_x)
    born = 10 * np.log10(sys.cal_snr())
    
    all_born.append(born)
```

```python
all_rms = []
all_rfocus = []
all_csm = []
all_perfect = []

for i in range(n_loop):
    if i%100==0:
        print("Computing loop ", i)
    sys.gen_channels(is_los=is_los)
    
    # sending samples and collect observation:
    X = np.zeros(shape=(sys.ris.N, N_samples))
    y = np.zeros(shape=(N_samples, 1))

    for j in range(N_samples):
        # generate a random ris setting
        sys.ris.set_random_binary_phase_shifts()
        x = sys.ris.vector_x
        X[:, j] = x.flatten()
        y[j] = sys.cal_snr() - sys.c_true
    
    # RMS
    sys.ris.vector_x = opt.RMS(X, y)
    rms = 10 * np.log10(sys.cal_snr())
    # RFocus
    sys.ris.vector_x = opt.RFocus(X, y)
    rfocus = 10 * np.log10(sys.cal_snr())
    # Conditional Sample Mean
    sys.ris.vector_x = opt.CSM(X, y)
    csm = 10 * np.log10(sys.cal_snr())
    # Perfect_csi
    sys.ris.theta, sys.ris.vector_x = opt.perfect_csi(h_0=sys.h_0, h_=sys.h)
    perfect_csi = 10 * np.log10(sys.cal_snr())
    
    all_rms.append(rms)
    all_rfocus.append(rfocus)
    all_csm.append(csm)
    all_perfect.append(perfect_csi)
```

```python
all_gcsm = []

for i in range(n_loop):
    if i%100==0:
        print("Computing loop ", i)
    sys.gen_channels(is_los=is_los)
    sys.run_GCSM(N_samples=N_samples)
    gcsm = 10 * np.log10(sys.cal_snr())

    all_gcsm.append(gcsm)
```

```python
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from matplotlib.patches import ConnectionPatch, Rectangle

def plot_multiple_cdfs_with_zoom(data_lists, labels):
    colors = ['#B71C1C', '#039BE5', '#26A69A', '#FF6F00', '#616161', '#BA68C8', '#EC407A']
    fig, ax = plt.subplots(figsize=(3.5, 2.625), dpi=500)

    # Font setup
    plt.rcParams['font.family'] = 'Helvetica'
    plt.rcParams['mathtext.fontset'] = 'dejavuserif'
    plt.rcParams['text.usetex'] = False
    plt.rcParams['font.size'] = 7
    plt.rcParams['legend.fontsize'] = 7
    plt.rcParams['xtick.minor.visible'] = False
    plt.rcParams['ytick.minor.visible'] = False
    plt.rcParams['xtick.direction'] = 'in'
    plt.rcParams['ytick.direction'] = 'in'
    plt.rcParams['xtick.major.width'] = 0.3
    plt.rcParams['ytick.major.width'] = 0.3
    plt.rcParams['axes.linewidth'] = 0.3

    # Plot main CDFs
    for i, x in enumerate(data_lists):
        sorted_x = np.sort(x)
        y = np.arange(1, len(x) + 1) / len(x)
        if labels[i] == 'BORN':
            ax.plot(sorted_x, y, drawstyle='steps-post', label=labels[i], ls='-.', color=colors[i], linewidth=1.0, alpha=0.8)
        elif labels[i] == 'Optimal':
            ax.plot(sorted_x, y, drawstyle='steps-post', label=labels[i], ls=':', color=colors[i], linewidth=1.0, alpha=0.8)
        else:
            ax.plot(sorted_x, y, drawstyle='steps-post', label=labels[i],
                color=colors[i], linewidth=1.0)

    ax.set_ylabel('CDF')
    ax.set_xlabel('Received SNR (dB)')
    ax.set_xlim([14, 30])
    ax.set_xticks([14, 18, 22, 26, 30])
    ax.set_ylim([0, 1])
    leg = ax.legend(frameon=False, edgecolor='black', framealpha=0.6)
    leg.get_frame().set_linewidth(0.4)
    ax.grid(color='gainsboro', linestyle='--', linewidth=0.2)

    # Create inset (zoom-in box)
    axins = inset_axes(
    ax,
    width="25%",         # width relative to main axes
    height="25%",        # height relative to main axes
    bbox_to_anchor=(0.3, -0.25, 1, 1),  # (x0, y0, width, height) in axes fraction
    bbox_transform=ax.transAxes,        # interpret coords relative to the big plot
    loc='center'                         # anchor point of inset
    )   

    # Plot same curves inside inset
    for i, x in enumerate(data_lists):
        sorted_x = np.sort(x)
        y = np.arange(1, len(x) + 1) / len(x)
        if labels[i] == 'BORN':
            axins.plot(sorted_x, y, drawstyle='steps-post', label=labels[i], ls='-.', color=colors[i], linewidth=1.0)
        elif labels[i] == 'Optimal':
            axins.plot(sorted_x, y, drawstyle='steps-post', label=labels[i], ls=':', color=colors[i], linewidth=1.0)
        else:
            axins.plot(sorted_x, y, drawstyle='steps-post', label=labels[i],
                color=colors[i], linewidth=1.0)

    # Set zoom limits
    x1, x2, y1, y2 = 23, 24, 0.5, 0.6
    axins.set_xlim(x1, x2)
    axins.set_ylim(y1, y2)
    axins.set_xticks([23, 23.5, 24])
    axins.set_yticks([0.5, 0.55,  0.6])
    axins.tick_params(axis='both', which='both', length=2, width=0.2, labelsize=4)
    axins.grid(color='gainsboro', linestyle='--', linewidth=0.2)

    rect = Rectangle((x1, y1), x2-x1, y2-y1,
                 fill=False, edgecolor="0.3", lw=0.4)
    ax.add_patch(rect)

    inset_corner = (0, 1)   # corner in Axes coordinates
    zoom_corner  = (x1, y1) # corner in data coordinates
    # Create connection line
    con = ConnectionPatch(xyA=inset_corner, coordsA=axins.transAxes,
                        xyB=zoom_corner, coordsB=ax.transData,
                        color="0.3", lw=0.4)
    ax.add_artist(con)

    inset_corner = (1, 1)
    zoom_corner  = (x2, y1)
    # Create connection line
    con = ConnectionPatch(xyA=inset_corner, coordsA=axins.transAxes,
                        xyB=zoom_corner, coordsB=ax.transData,
                        color="0.3", lw=0.4)
    ax.add_artist(con)

    plt.savefig(figure_name, dpi=500, bbox_inches='tight', pad_inches=0.05)
    plt.show()

plot_multiple_cdfs_with_zoom([all_born, all_perfect, all_gcsm, all_csm, all_rfocus, all_rms], labels=['BORN', 'Optimal', 'GCSM', 'CSM', 'RFocus', 'RMS'])
```
